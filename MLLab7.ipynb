```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd

```

```python
# 1. Load and prepare the data
# Load the Iris dataset, a classic multi-class classification dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target # Target labels (0, 1, 2 for the three types of iris flower)

# Split the data into training and testing sets (75% train, 25% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

```

```python
# 2. Initialize the SVM classifier
# Use the Support Vector Classifier (SVC) class from scikit-learn
# The 'linear' kernel is specified for this example, a simple yet powerful option
svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)

# 3. Train the model
# Fit the classifier to the training data
svm_classifier.fit(X_train, y_train)

# 4. Make predictions
# Use the trained model to predict the class labels for the test set
y_pred = svm_classifier.predict(X_test)

```

```python
# 5. Evaluate the model
# Calculate the accuracy of the model on the test data
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# You can also print a full classification report
from sklearn.metrics import classification_report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```
